```python
import random
print( "Rock vs Paper , paper  wins \n")
print( "Rock vs Scissors , rock  wins \n")
print( "Scissors vs Paper , Scissors wins \n")

while True:
    print("Enter your choice : 1 - Rock \n 2 - Paper \n 3 - Scissors \n")
    ch=int(input("Enter your choice :"))
           
    if ch == 1:
        ch_is= 'Rock'
    elif ch == 2:
        ch_is= 'Paper'
    else:
        ch_is= 'Scissors'
    print('User choice is \n',ch_is)
    print('Now Computers Turn')
    comp_ch = random.randint(1,3)

    while comp_ch == ch:
        comp_ch = random.randint(1,3)
    if comp_ch == 1:
        comp_ch_is = 'rocK'
    elif comp_ch == 2:
        comp_ch_is = 'papeR'
    else:
        comp_ch_is= 'scissoR'
    print("Computer choice is \n", comp_ch_is)
    print(ch_is,'Vs',comp_ch_is)

    if ch == comp_ch:
        print('Draw',end="")
        result="DRAW"

    if (ch==1 and comp_ch==2):
        print('paper wins =>',end="")
        result='PAPER'
    elif (ch==2 and comp_ch==1):
        print('paper wins =>',end="")
        result='Paper'

    if (ch==1 and comp_ch==3):
        print('Rock wins =>\n',end= "")
        result='Rock'
    elif (ch==3 and comp_ch==1):
        print('Rock wins =>\n',end= "")
        result='Rock'

    if (ch==2 and comp_ch==3):
        print('Scissors wins =>',end="")
        result='scissor'
    elif (ch==3 and comp_ch==2):
        print('Scissors wins =>',end="")
        result='Rock'

    if result == 'DRAW':
        print("  tie")
    if result == ch_is:
        print(" User wins")
    else:
        print("Computer wins ")
    print("Do you want to play again type y for yes and n for no ")

    ans = input().lower
    if ans =='n':
        break

print("thanks for playing")
```
